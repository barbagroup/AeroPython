```python
from math import * #operator
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.cm as cm
```

```python
N=100                          #Define number of points in array
x_low,x_high=-20.0, 20.0        #X Bounds
y_low,y_high=-20.0, 20.0        #Y Bounds
x=np.linspace(x_low,x_high,N) #X Array
y=np.linspace(y_low,y_high,N) #Y Array
print 'x=',x
print 'y=',y
X,Y=np.meshgrid(x,y)          #Mesh Grid
```

```python
%matplotlib inline
#plot
size=10
plt.figure(figsize=(size,(y_high-y_low)/(x_high-x_low)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(x_low,x_high)
plt.ylim(y_low,y_high)
plt.scatter(X,Y,s=10,color='#CD2305',marker='o',linewidth=0)
```

```python
sigma=7.5  #source strength
x_source,y_source=0.0,0.0
#velocity field on the mesh grid (strength, sigma)
u_source=sigma/(2*pi)*(X-x_source)/((X-x_source)**2+(Y-y_source)**2)
v_source=sigma/(2*pi)*(Y-y_source)/((X-x_source)**2+(Y-y_source)**2)
```

```python
#plot streamlines
size=10
plt.figure(figsize=(size,(y_high-y_low)/(x_high-x_low)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(x_low,x_high)
plt.ylim(y_low,y_high)
plt.streamplot(X,Y,u_source,v_source,density=2,linewidth=1,arrowsize=2,arrowstyle='->')
plt.scatter(x_source,y_source,color='#CD2305',s=80,marker='o',linewidth=0);
```

```python
strength_sink = -25.0                     # strength of the sink
x_sink, y_sink = 10.0, 10.0                # location of the sink

# computes the velocity on the mesh grid
u_sink = strength_sink/(2*pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
```

```python
# plots the streamlines
size = 10
plt.figure(figsize=(size, (y_high-y_low)/(x_high-x_low)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_low, x_high)
plt.ylim(y_low, y_high)
plt.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_sink, y_sink, color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
#Source-Sink Pair (Superposition)
u_pair=u_source+u_sink
v_pair=v_source+v_sink

#plot streamlines of pair
size=10
plt.figure(figsize=(size,(y_high-y_low)/(x_high-x_low)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_low, x_high)
plt.ylim(y_low, y_high)
plt.streamplot(X, Y, u_pair, v_pair, density=2.0, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter([x_source, x_sink], [y_source, y_sink], 
            color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
#Potential Flow Contour

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = mlab.bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
Z2 = mlab.bivariate_normal(X, Y, 1.5, 0.5, 1, 1)
# difference of Gaussians
Z = 10.0 * (Z2 - Z1)

# Create a simple contour plot with labels using default colors.  The
# inline argument to clabel will control whether the labels are draw
# over the line segments of the contour, removing the lines beneath
# the label
size=10
plt.figure(figsize=(size,(y_high-y_low)/(x_high-x_low)*size))
CS = plt.contour(X,Y,Z)
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Contour Plot')


# contour labels can be placed manually by providing list of positions
# (in data coordinate). See ginput_manual_clabel.py for interactive
# placement.
plt.figure()
CS = plt.contour(X, Y, Z)
manual_locations = [(-1, -1.4), (-0.62, -0.7), (-2, 0.5), (1.7, 1.2), (2.0, 1.4), (2.4, 1.7)]
plt.clabel(CS, inline=1, fontsize=10, manual=manual_locations)
plt.title('labels at selected locations')
```
