```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50                                # number of points in each direction
x_start, x_end = -2.0, 2.0            # boundaries in the x-direction
y_start, y_end = -1.0, 1.0            # boundaries in the y-direction
x = np.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = np.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates

# print 'x = ', x
# print 'y = ', y

X, Y = np.meshgrid(x, y)              # generates a mesh grid

%matplotlib inline

# plots the grid of points
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.scatter(X, Y, s=10, color='#CD2305', marker='o', linewidth=0)
```

```python
strength_source = 5.0                      # source strength
strength_sink = -5.0                     # strength of the sink
x_source, y_source = -1.0, 0.0             # location of the source
x_sink, y_sink = 1.0, 0.0                # location of the sink

r_source = ((X-x_source)**2 + (Y-y_source)**2)
r_sink = ((X-x_sink)**2 + (Y-y_sink)**2)

# computes the velocity field on the mesh grid
u_source = strength_source/(2*pi) * (X-x_source)/r_source
v_source = strength_source/(2*pi) * (Y-y_source)/r_source

# computes the velocity on the mesh grid
u_sink = strength_sink/(2*pi) * (X-x_sink)/r_sink
v_sink = strength_sink/(2*pi) * (Y-y_sink)/r_sink

# combines the source and sink
u_sourcesink = u_source + u_sink
v_sourcesink = v_source + v_sink

# calculcate potential
phi_source = strength_source/(2*pi) * np.log(r_source)
phi_sink = strength_sink/(2*pi) * np.log(r_sink)
phi = phi_source + phi_sink

# plotting the streamlines
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_sourcesink, v_sourcesink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter([x_source,x_sink], [y_source,y_sink], color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
# plotting the potential
size = 10
plt.figure(figsize=(size, 0.75 * size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
cs = plt.contourf(X, Y, phi, levels = np.linspace(-8,8,100))
plt.colorbar(cs, orientation='horizontal', format="%.2f")
```

```python
# Finding the mass flux
rho = 1000.
uflux = rho*u_source
vflux = rho*u_source

print(uflux[x_source,y_source],vflux[x_source,y_source])
```
