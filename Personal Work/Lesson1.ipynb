```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer
```

```python
start = timer()
N = 500                               # number of points in each direction
x_start, x_end = -2.0, 2.0            # boundaries in the x-direction
y_start, y_end = -1.0, 1.0            # boundaries in the y-direction
x = np.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = np.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates

# print 'x = ', x
# print 'y = ', y

X, Y = np.meshgrid(x, y)              # generates a mesh grid

vel = np.zeros((np.size(X,0),np.size(X,1),2))
phi = np.zeros_like(X)

%matplotlib inline
```

```python
strength_source = 5.0                      # source strength
strength_sink = -5.0                  # strength of the sink
x_source, y_source = -1.0, 0.0             # location of the source
x_sink, y_sink = 1.0, 0.0                # location of the sink

r_source = np.sqrt((X-x_source)**2 + (Y-y_source)**2)
r_sink = np.sqrt((X-x_sink)**2 + (Y-y_sink)**2)

calcstart = timer()

# more efficient velocity calculations
vel[:,:,0] = (1/(2*pi))*((strength_source * (X-x_source)/(r_source**2)) + \
                         (strength_sink * (X-x_sink)/(r_sink**2)))
vel[:,:,1] = (1/(2*pi))*((strength_source * (Y-y_source)/(r_source**2)) + \
                         (strength_sink * (Y-y_sink)/(r_sink**2)))

# calculcate potential
phi = (1/(2*pi))*(strength_source * np.log(r_source) + \
                  strength_sink * np.log(r_sink))

calcend = timer()

# plotting the streamlines
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, vel[:,:,0], vel[:,:,1], density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter([x_source,x_sink], [y_source,y_sink], color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
# plotting the potential
size = 10
plt.figure(figsize=(size, 0.75 * size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
cs = plt.contour(X, Y, phi, levels = np.linspace(np.min(phi),\
                                                  np.max(phi),100))
plt.colorbar(cs, orientation='horizontal', format="%.2f")
```

```python
%timeit
# Find Q for the source and sink
u_rsource = np.sqrt(vel[x_source,y_source,0]**2. + vel[x_source,y_source,1]**2.)
u_rsink = np.sqrt(vel[x_sink,y_sink,0]**2. + vel[x_sink,y_sink,1]**2.)
Q_source = u_rsource * (2. * pi)
Q_sink = u_rsink * (2. * pi)
end = timer()

print(Q_source,Q_sink)
print('Total Time = ' + repr(end - start) + ' seconds')
print('Math Time = ' + repr(calcend - calcstart) + ' seconds')

# As N -> infinity, Q_sink -> Q_source
```
