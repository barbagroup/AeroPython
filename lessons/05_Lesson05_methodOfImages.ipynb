# Aerodyanmic interferences  Method of images

In the is notebook, we will introduce the method of images to compute the streamlines of various singularities in the proximity of a wall. The velocity field induced by a singularity will be affected by the presence of a wall.

In addition to that, this notebook also gives us the opportunity to introduce the notion of classes in Python.

As usual, we start by importing the libraries and creating the computational grid.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from IPython.core.display import clear_output
```

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

## Source near a plane wall

We start here by illustrating the distorsion of the flow emanating from a source, located at $y=y_\text{source}$, in the presence of a wall ($y=0$). The boundary condition at the wall requires the flow to be tangent, i.e. $v=0$, which can be perform by observing that the effect of the wall on the source is identical as placing an other source (an image) of equal strength, located at $y=-y_\text{source}$.

Ok, now it is time to play smart and avoid repeating code that is wasting our precious time! In a previous notebook, we already introduced th notion of function in Python, and now we are going to move forward with the creation of classes.

We define a `class` called Source that will contain information related to a source. A singularity such as a source is defined by its strength and its location in the domain. Therefore, our class `Source` will be initiated with three `attributes` defining our singularity:

* `strength`: the strength of the source.
* `x`: the loction of the source of the $x$-axis.
* `y`: the loction of the source of the $y$-axis.
What would we like to do after the definition of our source? We want to compute the velocity field, as well as the stream-function, induced by the source. So, in our class `Source`, we have to implement two methods (functions included in a class): one to compute the velocity (a function called `velocity`) and the other one to calculate the stream-function (a function called `streamFunction`).

Let have a detailed look of the methods:

* method `velocity`: is a function used to calculate the velocity on the mesh (`X`,`Y`) induced by a source. Therefore, our method will have two arguments, `X` and `Y`, and generates two new attributes of our source: `u` and `v`, the velocity components induced by the source.
* method `streamFunction`: is a function used to calculate the stream-function on the mesh (`X`,`Y`) induced by a source. Therefore, our method will have two arguments, `X` and `Y`, and generates one new attribute of our source: `psi`, the stream-function induced by the source.

```python
class Source:
    def __init__(self,strength,x,y):
        self.strength = strength
        self.x,self.y = x,y
    # get the velocity field
    def velocity(self,X,Y):
        self.u = self.strength/(2*pi)*(X-self.x)/((X-self.x)**2+(Y-self.y)**2)
        self.v = self.strength/(2*pi)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)
    # get the stream-function
    def streamFunction(self,X,Y):
        self.psi = self.strength/(2*pi)*np.arctan2((Y-self.y),(X-self.x))
```

We define an object called `source` as an instantiation of the class `Source`, defining the strength and the location of the singularity. Then we call our two methods to compute the velocity field and the stream-function on the mesh (`X`,`Y`) induced by the singularity.

The Cartesian velocity components are given by:

$$u = \frac{\sigma_\text{source}}{2\pi}\frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

$$v = \frac{\sigma_\text{source}}{2\pi}\frac{y-y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

and the stream-function is defined by:

$$\psi = \frac{\sigma_\text{source}}{2\pi}\arctan\left(\frac{y-y_\text{source}}{x-x_\text{source}}\right)$$

```python
strengthSource = 1.0
xSource,ySource = 0.0,0.5
source = Source(strengthSource,xSource,ySource)
source.velocity(X,Y)
source.streamFunction(X,Y)
```

As said earlier, the presence of the wall is replaced by the addition of an image  of the source, located on the other side of the wall. The image is also an object of the class `Source`, same strength but different location. One can now see the efficiency of having created a class.

```python
sourceImage = Source(strengthSource,xSource,-ySource)
sourceImage.velocity(X,Y)
sourceImage.streamFunction(X,Y)
```

Using the now well-known principle of superposition, we can compute the streamlines induced by a source in the vicinity of a wall.

The superimposition of the two sources leads to the following velocity field:

$$u = \frac{\sigma_\text{source}}{2\pi} \left( \frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2} + \frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y+y_\text{source}\right)^2} \right)$$

$$v = \frac{\sigma_\text{source}}{2\pi} \left( \frac{y-y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2} + \frac{y+y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y+y_\text{source}\right)^2} \right)$$

and the following stream-function:

$$\psi = \frac{\sigma_\text{source}}{2\pi} \left( \arctan\left(\frac{y-y_\text{source}}{x-x_\text{source}}\right) + \arctan\left(\frac{y+y_\text{source}}{x-x_\text{source}}\right) \right)$$

```python
# Superimposition of the source and its image
u = source.u + sourceImage.u
v = source.v + sourceImage.v
psi = source.psi + sourceImage.psi

# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(source.x,source.y,c='r',s=80,marker='o')
plt.scatter(sourceImage.x,sourceImage.y,c='r',s=80,marker='*')
plt.axhline(0.0,color='k',linestyle='--',linewidth=4)
```

## Vortex near a plane wall

In a similar manner, we can write an other class `Vortex`, instantiated by a strength and a location, to create a singularities of type vortex. The class possesses two methods: one to compute the velocity in the domain induced by the vortex and the other one to calculate the stream-function.

```python
class Vortex:
    def __init__(self,strength,x,y):
        self.strength = strength
        self.x,self.y = x,y
        
    # get velocity field
    def velocity(self,X,Y):
        self.u = +self.strength/(2*pi)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)
        self.v = -self.strength/(2*pi)*(X-self.x)/((X-self.x)**2+(Y-self.y)**2)
        
    # get streamfunction
    def streamFunction(self,X,Y):
        self.psi = -self.strength/(4*pi)*np.log((X-self.x)**2+(Y-self.y)**2)
```

It is now very quick to create two object of type `Vortex` for the vortex and its image.

```python
strengthVortex = 1.0
xVortex,yVortex = 0.0,0.5

vortex = Vortex(strengthVortex,xVortex,yVortex)
vortex.velocity(X,Y)
vortex.streamFunction(X,Y)

vortexImage = Vortex(-strengthVortex,xVortex,-yVortex)
vortexImage.velocity(X,Y)
vortexImage.streamFunction(X,Y)
```

Applying the principle of superposition, we are able to plot the streamlines induced by the presence of a vortex near a wall.

```python
# Superimposition of the vortex and its image
u = vortex.u + vortexImage.u
v = vortex.v + vortexImage.v
psi = vortex.psi + vortexImage.psi

# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(vortex.x,vortex.y,c='r',s=80,marker='o')
plt.scatter(vortexImage.x,vortexImage.y,c='r',s=80,marker='*')
plt.axhline(0.0,color='k',linestyle='--',linewidth=4)
```

## Motion of a vortex pair near the ground

```python
strengthVortex = 1.0
xVortex1,yVortex1 = -0.1,0.5
xVortex2,yVortex2 = 0.1,0.5

vortex1 = Vortex(+strengthVortex,xVortex1,yVortex1)
vortex2 = Vortex(-strengthVortex,xVortex2,yVortex2)

vortex1.velocity(X,Y)
vortex1.streamFunction(X,Y)
vortex2.velocity(X,Y)
vortex2.streamFunction(X,Y)

vortexImage1 = Vortex(-strengthVortex,xVortex1,-yVortex1)
vortexImage2 = Vortex(+strengthVortex,xVortex2,-yVortex2)

vortexImage1.velocity(X,Y)
vortexImage1.streamFunction(X,Y)
vortexImage2.velocity(X,Y)
vortexImage2.streamFunction(X,Y)
```

```python
# Superimposition of the vortex pair and its image
u = vortex1.u + vortex2.u + vortexImage1.u + vortexImage2.u
v = vortex1.v + vortex2.v + vortexImage1.v + vortexImage2.v
psi = vortex1.psi + vortex2.psi + vortexImage1.psi + vortexImage2.psi

# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(vortex1.x,vortex1.y,c='r',s=80,marker='o')
plt.scatter(vortex2.x,vortex2.y,c='g',s=80,marker='o')
plt.scatter(vortexImage1.x,vortexImage1.y,c='r',s=80,marker='*')
plt.scatter(vortexImage2.x,vortexImage2.y,c='g',s=80,marker='*')
plt.axhline(0.0,color='k',linestyle='--',linewidth=4)
```

## Doublet near a plane wall parallel to a uniform flow

```python
class Freestream:
    def __init__(self,U,alpha):
        self.U = U
        self.alpha = alpha*pi/180
        
    # get the velocity
    def velocity(self):
        self.u,self.v = self.U*cos(self.alpha),self.U*sin(self.alpha)
        
    # get stream function
    def streamFunction(self,X,Y):
        self.psi = self.U*cos(self.alpha)*Y - self.U*sin(self.alpha)*X
```

```python
class Doublet:
    def __init__(self,strength,x,y):
        self.strength = strength
        self.x,self.y = x,y
        
    # get velocity field
    def velocity(self,X,Y):
        self.u = -self.strength/(2*pi)*\
                ((X-self.x)**2-(Y-self.y)**2)/((X-self.x)**2+(Y-self.y)**2)**2
        self.v = -self.strength/(2*pi)*\
                2*(X-self.x)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)**2
            
    # get stream function
    def streamFunction(self,X,Y):
        self.psi = -self.strength/(2*pi)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)
```

```python
Uinf,alpha = 1.0,0.0
freestream = Freestream(Uinf,alpha)
freestream.velocity()
freestream.streamFunction(X,Y)

strengthDoublet = 1.0
xDoublet,yDoublet = 0.0,0.3
doublet = Doublet(strengthDoublet,xDoublet,yDoublet)
doublet.velocity(X,Y)
doublet.streamFunction(X,Y)

doubletImage = Doublet(strengthDoublet,xDoublet,-yDoublet)
doubletImage.velocity(X,Y)
doubletImage.streamFunction(X,Y)
```

```python
# Superimposition of the doublet and its image to the uniform flow
u = freestream.u + doublet.u + doubletImage.u
v = freestream.v + doublet.v + doubletImage.v
psi = freestream.psi + doublet.psi + doubletImage.psi

# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(doublet.x,doublet.y,c='r',s=80,marker='o')
plt.scatter(doubletImage.x,doubletImage.y,c='r',s=80,marker='*')
plt.axhline(0.0,color='k',linestyle='--',linewidth=4)
```
