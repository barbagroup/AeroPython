## Overview

The theoretical basis for classical Aerodynamics is *potential-flow theory*, a mathematical apparatus that was the life-blood of aerodynamics through its growth to maturity during most of the 20th century.

It all starts with making some simplifications: 

* the flow is steady;
* the velocity remains smaller than the speed of sound (incompressible flow);
* the fluid has no internal friction, i.e., is inviscid; and
* it has no vorticity (fluid particles are not rotating). 

This sounds like a lot of simplifications, does it not? It turns out, a big chunk of aerodynamics can be approximated this way! Viscous effects are normally confined to a very thin boundary layer (and we can correct potential theory to account for that), and many flows are effectively irrotational (except for isolated points, lines or sheets). And finally, most applications that we are interested in do remain largely sub-sonic.

Potential-flow theory has a very pleasant mathematical quality: it is *linear*. This means that the principle of superposition applies, and we can construct new solutions by adding known solutions. 

In this notebook, you will learn two elementary potential-flow solutions: the **source**, and the **sink**. And, guess what, you will add them together to make a *new* solution called a **source-sink pair**.

The objective is to visualize the streamlines induced by a singularity such as a source or a sink. To do that, we need to define a two-dimensional mesh where the velocity components will be computed.

We would like to define an evenly spaced Cartesian grid of points within a spatial domain that is 4 units of length wide in the $x$-direction and 2 units of length wide in the $y$-direction, i.e. $x,y\in\left[-2,2\right],\left[-1,1\right]$.

We define `N` the number of points  we want in the two directions, as well as the computational boundaries `xStart`,`xEnd`,`yStart` and `yEnd`. Then, we use the `numpy` function `linspace` to create two 1D arrays that contain the evenly spaced values of $x$ and $y$. The last line calls an other function of numpy `meshgrid` to generate a grid containing the coordinates of points where the numerical solution will be calculated.

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
print 'x = ',x
print 'y = ',y
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

Now the mesh grid has been generated, it is time to visualize it with the module `pyplot` from the library `matplotlib` using the alias `plt` and the function `scatter`.

```python
# plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.scatter(X,Y,s=8,c='b',marker='o')
```

## $\bullet$ Source flow

Let consider a two-dimensional incompressible flow where all the streamlines emanate from a single point as straight lines and where the radial velocity is inversionally proportional to the distance from the point and the tangential velocity is zero. Such a flow is called a source. In a cylindrical coordinate system $\left(r,\theta\right)$, the velocity components are :

$$u_r\left(r,\theta\right) = \frac{\sigma_\text{source}}{2\pi r} \quad \text{,} \quad u_\theta\left(r,\theta\right)=0$$

where $\sigma_\text{source}$ denotes the source strength.

The stream function $\psi$ is obtained as follow:

$$\frac{1}{r}\frac{\partial\psi}{\partial\theta} = u_r \quad \text{,} \quad -\frac{\partial\psi}{\partial r} = u_\theta$$

leading to

$$\psi = \frac{\sigma_\text{source}}{2\pi}\theta + \text{constant}$$

In practical problems, we are more interested in the velocity components that are obtained by differentiation of the stream function, so that the constant can be dropped.

In Cartesian coordinates, the velocity field $\left(u,v\right)$ at position $\left(x,y\right)$ induced by the source $\sigma_\text{source}$ located at $\left(x_\text{source},y_\text{source}\right)$ are given by:

$$u = \frac{\partial\psi}{\partial y} = \frac{\sigma_\text{source}}{2\pi}\frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

and

$$v = -\frac{\partial\psi}{\partial x} = \frac{\sigma_\text{source}}{2\pi}\frac{y-y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

```python
source = 5.0
xSource, ySource = -1.0,0.0

uSource = np.empty((N,N),dtype=float)
vSource = np.empty((N,N),dtype=float)

for i in range(N):
    for j in range(N):
        uSource[i,j] = source/(2*pi)\
                    *(X[i,j]-xSource)/((X[i,j]-xSource)**2+(Y[i,j]-ySource)**2)

        vSource[i,j] = source/(2*pi)\
                    *(Y[i,j]-ySource)/((X[i,j]-xSource)**2+(Y[i,j]-ySource)**2)
```

```python
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uSource,vSource,density=2.0,linewidth=1,arrowsize=2,arrowstyle='->')
plt.scatter(xSource,ySource,c='r',s=80,marker='o')
```

## $\bullet$ Sink flow

In the previous paragraph, the strength $\sigma$ of the singularity was chosen to be positive. A source with a negative strength is called a sink. Instead of emanating from a single point, the straight streamlines are converging to a single point.

The velocity field induced by a sink is similar the one induced by a source, except that the singularity strength is negative ($\sigma_\text{sink} < 0$). And the Python code requires very few modifications.

```python
strengthSink = -5.0                          # sink strength
xSink,ySink = 1.0,0.0                        # location of the sink

uSink = np.empty((N,N),dtype=float)
vSink = np.empty((N,N),dtype=float)

for i in range(N):
    for j in range(N):
        uSink[i,j] = strengthSink/(2*pi)\
                    *(X[i,j]-xSink)/((X[i,j]-xSink)**2+(Y[i,j]-ySink)**2)

        vSink[i,j] = strengthSink/(2*pi)\
                    *(Y[i,j]-ySink)/((X[i,j]-xSink)**2+(Y[i,j]-ySink)**2)
```

```python
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uSink,vSink,density=2.0,linewidth=1,arrowsize=2,arrowstyle='->')
plt.scatter(xSink,ySink,c='r',s=80,marker='o')
```

## $\bullet$ Source-sink pair

```python
uPair = np.empty_like(uSource)
vPair = np.empty_like(vSource)

uPair[:,:] = uSource[:,:] + uSink[:,:]
vPair[:,:] = vSource[:,:] + vSink[:,:]

size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uPair,vPair,density=2.0,linewidth=1,arrowsize=2,arrowstyle='->')
plt.scatter([xSource,xSink],[ySource,ySink],c='r',s=80,marker='o')
```
