To avoid edge effects, a large number of vortices are distributed over the domain of interest. In the meanwhile, zooming in and plotting the center part. Doing so also slows down the computation significantly. A way to work around is that only counting the influence of all the vortices within the region of plotting. The solution is guaranteed by the nature of linearity of potential flow.

When using odd number of node (even number of intervals) and odd number of vortices at the same time, on a 2D axisymmetric flow field, could result in "dividing by zero". Make sure at least one of them is even number. 

```python
import numpy
import math
from matplotlib import pyplot

N = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -0.5, 0.5
x_start_p, x_end_p = -2.0, 2.0
y_start_p, y_end_p = -0.5, 0.5
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x, y)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
        """Returns the velocity field generated by a vortex.
        
        Arguments
        ---------
        strength -- strength of the vortex.
        xv, yv -- coordinates of the vortex.
        X, Y -- mesh grid.
        """
        u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
        v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)

        return u, v
```

```python
def get_stream_function_vortex(strength, xv, yv, X, Y):
    """Returns the stream-function generated by a vortex.
            
    Arguments
    ---------
    strength -- strength of the vortex.
    xv, yv -- coordinates of the vortex.
    X, Y -- mesh grid.
    """
    psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)

    return psi
```

```python
nv         = 1100 #number of vortex 
Vstrength  = 5.0
x_source   = numpy.linspace(100*x_start,100*x_end,nv)
y_source   = numpy.zeros(nv, dtype=float)
```

```python
u_source   = numpy.zeros((nv,N, N), dtype=float)
v_source   = numpy.zeros((nv,N, N), dtype=float)
psi_source = numpy.zeros((nv,N, N), dtype=float)

u_total    = numpy.zeros((N, N), dtype=float)
v_total    = numpy.zeros((N, N), dtype=float)
psi_total  = numpy.zeros((N, N), dtype=float)
```

```python
for i in range(0,nv):
    u_source[i,:,:], v_source[i,:,:] = get_velocity_vortex(Vstrength,x_source[i], y_source[i], X, Y)
    psi_source[i,:,:] = get_stream_function_vortex(Vstrength,x_source[i],y_source[i], X, Y)
```

```python
for i in range(0,nv):
    u_total += u_source[i,:,:]
    v_total += v_source[i,:,:]
    psi_total += psi_source[i,:,:]
```

```python
u_source   = numpy.zeros((N, N), dtype=float)
v_source   = numpy.zeros((N, N), dtype=float)
```

```python
%matplotlib inline

size = 10
pyplot.figure(1,figsize=(size, (y_end_p-y_start_p)/(x_end_p-x_start_p)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start_p, x_end_p)
pyplot.ylim(y_start_p, y_end_p)
pyplot.streamplot(X, Y, u_total, v_total, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o')
```

```python
def get_velocity_infvortex(strength, a, X, Y):
        """Returns the velocity field generated by a vortex.
        
        Arguments
        ---------
        strength -- strength of the vortex.
        a -- spacing between two adjacent vortices
        xv, yv -- coordinates of the vortex.
        X, Y -- mesh grid.
        """
        u = + strength/(2*a) * numpy.sinh(2*math.pi*Y/a)/(numpy.cosh(2*math.pi*Y/a)-numpy.cos(2*math.pi*X/a))
        v = - strength/(2*a) * numpy.sin(2*math.pi*X/a)/(numpy.cosh(2*math.pi*Y/a)-numpy.cos(2*math.pi*X/a))

        return u, v
```

```python
u_infsource, v_infsource = get_velocity_infvortex(Vstrength,0.4,X,Y)
```

```python
size = 10
pyplot.figure(2,figsize=(size, (y_end_p-y_start_p)/(x_end_p-x_start_p)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start_p, x_end_p)
pyplot.ylim(y_start_p, y_end_p)
pyplot.streamplot(X, Y, u_infsource, v_infsource, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
```
