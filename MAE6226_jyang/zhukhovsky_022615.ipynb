```python
import numpy
from matplotlib import pyplot
```

```python
Ntheta = 145
Nr = 100
r1 = 1.15
r2 = 5
xc = -0.15
yc = 0
```

```python
thetagrid = numpy.linspace(0,2*numpy.pi,Ntheta)
```

```python
rgrid = numpy.linspace(r1,r2,Nr)
```

```python
X = numpy.zeros((Nr,Ntheta),dtype=float)
Y = numpy.zeros((Nr,Ntheta),dtype=float)
for i in range(0,Ntheta):
    for j in range(0,Nr):
        X[j][i] = rgrid[j]*numpy.sin(thetagrid[i])+xc
        Y[j][i] = rgrid[j]*numpy.cos(thetagrid[i])
        
xi = numpy.ndarray((Nr,Ntheta),dtype=complex)
for i in range(0,Ntheta):
    for j in range(0,Nr):
        sigma = X[j][i] + Y[j][i]*1j
        xi[j][i] = sigma + 1./sigma
```

```python
%matplotlib inline
size = 8
limit = 6
pyplot.figure(1,figsize=(size, size))
pyplot.xlim((-limit,limit))
pyplot.ylim((-limit,limit))
pyplot.suptitle('z-plane')
pyplot.scatter(X,Y, color='#CD2305', s=2,marker='o')

size = 8
limit = 6
pyplot.figure(2,figsize=(size, size))
pyplot.xlim((-limit,limit))
pyplot.ylim((-limit,limit))
pyplot.suptitle(r'$\xi$''-plane')
pyplot.scatter(xi.real,xi.imag, color='#CD2305', s=2,marker='o')
```
